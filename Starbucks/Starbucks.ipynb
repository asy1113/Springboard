# Cleaning the data

```python
df = df.drop(columns = ['Trans Fat (g) ', 'Saturated Fat (g)','Vitamin A (% DV) ','Vitamin C (% DV)',
                        ' Calcium (% DV) ', 'Iron (% DV) ', ' Dietary Fibre (g)'])
```

```python
df = df.rename(columns = {'Beverage_category':'beverage_cagegory','Beverage':'beverage','Milk':'milk',
                          'Size':'size','Calories':'calories',' Total Fat (g)': 'fat',' Sodium (mg)':'sodium',
                          ' Total Carbohydrates (g) ':'carb' , 'Cholesterol (mg)':'cholesterol',
                          ' Sugars (g)':'sugar', ' Protein (g) ':'protein', 'Caffeine (mg)':'caffeine'}) 


```

```python
# check any null value

df.isnull().sum()
```

```python
# there is no negative value in caffeine column

df[df.caffeine<0]
```

```python
df[df.caffeine.isnull()]

# null caffeine values are associated with Tazo tea drinks (caffeine = varies in tazo tea)
```

```python
# dropping missing caffeine info beverages 

df = df[df.caffeine>=0]
```

```python
# Filter Grande drink (easy to compare all drinks when they have the same size)

df = df[df['size']=='Grande']
print (f"There are {df.shape[0]} Grande drinks at Starbucks")
df.head()
```

# Relationships

```python
# the average value of each nutrient for each category of drinks

categ_avg = df.groupby('beverage_cagegory').mean()
categ_avg
```

```python
# Plot the distribution for all features =, and number of drinks (66 drinks with Grande size)
df.hist(layout=(2,4), figsize=(12,8))      
plt.subplots_adjust(hspace = 0.5);
```

```python
# Correlation between type of drink and its nutrient
plt.figure(figsize=(12,10))
sns.heatmap(categ_avg, cmap='Blues', annot=True)
```

```python
# Show a seaborn heatmap of correlations in starbucks nutrient
plt.subplots(figsize=(12,10))
sns.heatmap(df.corr(), annot=True);
```

```python
# Calories, sugar and carb has high correlation with each other

plt.scatter(data=df, x='calories', y='sugar')
plt.xlabel('Calories')
plt.ylabel('Sugar')
plt.title('Calories vs. Sugar');
```

```python
plt.scatter(data=df, x='calories', y='carb', alpha=1/2)
plt.xlabel('Calories')
plt.ylabel('Carb')
plt.title('Calories vs. Carb');
```

```python
plt.scatter(data=df, x='sugar', y='carb')
plt.xlabel('Sugar')
plt.ylabel('Carb')
plt.title('Sugar vs. Carb');
```

# Visualize nutrients in Starbucks drinks

```python
calories = df.sort_values(by='calories', ascending=False)
carb = df.sort_values(by='carb', ascending=False)
sugar = df.sort_values(by='sugar', ascending=False)
caffeine = df.sort_values(by='caffeine', ascending=False)
```

```python
caffeine.head(10)
```

```python
# Calories
top_10_high_calories = calories.head(10)

top_10_high_calories.plot(kind = 'scatter', x = 'calories', y = 'beverage', 
                        color='red', label='Calories')
plt.xlabel('Calories')
plt.ylabel('Beverage Name')
plt.title('Top 10 Starbucks high calories drinks', size=20)
plt.show();

# Carb
top_10_high_carb = carb.head(10)
top_10_high_carb.plot(kind = 'scatter', x = 'carb', y = 'beverage', 
                        color='orange', label='carb')
plt.xlabel('Carbohydrate')
plt.ylabel('Beverage Name')
plt.title('Top 10 Starbucks high carb drinks', size=20)
plt.show();

# Sugar
top_10_high_sugar = sugar.head(10)
top_10_high_sugar.plot(kind = 'scatter', x = 'sugar', y = 'beverage', 
                        color='purple', label='carb')
plt.xlabel('Sugar')
plt.ylabel('Beverage Name')
plt.title('Top 10 Starbucks high sugar drinks', size=20)
plt.show();

# Caffeine
top_10_high_caffeine = caffeine.head(10)
top_10_high_caffeine.plot(kind = 'scatter', x = 'caffeine', y = 'beverage', 
                        color='black', label='carb')
plt.xlabel('Caffeine')
plt.ylabel('Beverage Name')
plt.title('Top 10 Starbucks high caffeine drinks', size=20)
plt.show();
```
